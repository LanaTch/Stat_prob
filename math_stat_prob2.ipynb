### Задача 1
Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. 

Составьте закон распределения числа правильных ответов при простом угадывании. Найдите M(X), D(X).


```python
import numpy as np
import scipy.special as ss

""" вероятность правильно ответить на 1 вопрос из 4: p = 0.25 (все исходны равновероятны), 
    количество вопросов n = 5 (порядок бинома), x = [0, 1, 2, 3, 4, 5] (случайная величина - количество правильных ответов)
"""
n = 5

x_values = np.arange(6) 
print(f'значения, которые принимает случайная величина: {x_values}')

C = ss.binom(n, range(6)) # возвращает массив биномиальных коэффициентов

p = 0.25
x_prob = np.array([C[i]*p**i*(1-p)**(n-i) for i in range(6)])
print(f'вероятности, соответствующие значениям случайной величины: {x_prob}')

m = x_values.dot(x_prob)
print(f'математическое ожидание = {m}')

d = ((x_values - m) ** 2).dot(x_prob)
print(f'дисперсия = {d}')
```

### Задача 2
Вероятность попадания в цель при одном выстреле равна 0.01. 

Какова вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10?


```python
p = 0.01
n = 200

x_values = np.arange(5, 11) # от 5 до 10 попаданий 
print(f'значения, которые принимает случайная величина: {x_values}')

C = ss.binom(n, np.arange(5, 11)) # возвращает массив биномиальных коэффициентов

x_prob = np.array([C[i-5]*p**i*(1-p)**(n-i) for i in range(5, 11)]).sum()
print(f'вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10 = {x_prob}')
```

### Задача 3
В условиях предыдущей задачи сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, 

что хотя бы 10 раз будет совершено попадание?

Подсказка. Как посчитать P(X >= k) для величины X, имеющей распределение Пуассона? Здесь предстоит немножко покодить.


```python
from math import factorial

n = 1400
p = 0 # вероятность, что будет >= 10 попаданий
while p < 0.9: # вычисления повторяем пока не достигнем искомой вероятности
    q = 0 # вероятность, что будет < 10 попаданий. такую вероятность проще посчитать, чем искомую
    n += 1 # счетчик выстрелов
    for m in range(10): # m - количество попаданий от 0 до 9
        lambda_ = n*0.01
        q += (lambda_)**m/factorial(m)*np.exp(-lambda_) # в итоге будет сумма вероятностей попдания от 0 до 9 раз
    p = 1 - q # находим искомую вероятность

print(f'при количестве выстрелов n = {n}, вероятность попадания хотя бы 10 раз p = {p}')
```
